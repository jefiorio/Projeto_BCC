Para finalizar, nós resolvemos criar um gráfico de dispersão mostrando as vendas de 1977 a 2020.

```python
df = pd.read_csv("https://drive.google.com/u/1/uc?id=1Tc1C4zFRon0Lk6O5iYOvUyGAGuJ3-bX6&export=download")
df
x = df["Year"]
y = df['Total_Shipped']
plt.plot(x,y,'o')
```

Percebe-se que poucos jogos conseguiram ultrapassar a linha de 20 milhões de vendas. Também é perceptível que a quantidade jogos produzidos ao longo desses anos foi aumentando. Mas será que a qualidade deles acompanhou o crescimento?

Para analisar esse campo, nós começamos pela média das notas dos jogos lançados a cada ano.

```python
df = pd.read_csv("https://drive.google.com/u/1/uc?id=1Tc1C4zFRon0Lk6O5iYOvUyGAGuJ3-bX6&export=download")
df
df.groupby("Year")["Critic_Score"].mean()
```

Que tal uma média das notas de todos os jogos da planilha?

```python
df = pd.read_csv("https://drive.google.com/u/1/uc?id=1Tc1C4zFRon0Lk6O5iYOvUyGAGuJ3-bX6&export=download")

df["Critic_Score"].mean()
```

Para aprofundar mais a relação entre o passar dos anos e a qualidade dos jogos, nós fizemos uma correlação entre as notas de críticos especializados da indústria e o passar dos anos.

```python
df = pd.read_csv("https://drive.google.com/u/1/uc?id=1Tc1C4zFRon0Lk6O5iYOvUyGAGuJ3-bX6&export=download")
df
corr = df["Critic_Score"].corr(df["Year"])
print('%.3f'%corr)
```

Com isso, notamos que há uma correlação fraca, e portanto, pode-se dizer que a qualidade dos jogos não tendeu a crescer com o passar dos anos.

Também procuramos buscar uma relação entre as notas dos críticos e as notas dos consumidores dos jogos para saber se as opiniões são semelhantes.

```python
df = pd.read_csv("https://drive.google.com/u/1/uc?id=1Tc1C4zFRon0Lk6O5iYOvUyGAGuJ3-bX6&export=download")

corr = df["User_Score"].corr(df["Critic_Score"])
print('%.3f'%corr)
```

Há uma correlação moderada, portanto pode-se dizer que elas são semelhantes de certo modo.

Por fim, decidimos pegar o top 20 de jogos mais vendidos nas principais plataformas. Começando pelo PC!

```python
df = pd.read_csv("https://drive.google.com/u/1/uc?id=1Tc1C4zFRon0Lk6O5iYOvUyGAGuJ3-bX6&export=download")
df
df = df.query("Platform == 'PC'")
df.sort_values(by="Total_Shipped", inplace=True, ascending=False,  ignore_index=True)
df_rank20 = df.head(20)

df_rank20
```
